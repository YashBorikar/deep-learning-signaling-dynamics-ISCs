### Importing Libraries

```python
# Importing YOLOV8 library
from ultralytics import YOLO
```

### Training YOLO-V8 Small model

```python
# Instance of yolo-v8 small model
model_s = YOLO()
model_s = YOLO('yolov8s.pt')
```

##### Hyperparameters for model training

```python
data = '/content/drive/MyDrive/Dissertation/Data/TrainingDataset/data.yaml'
# Change directory of dataset
epochs = 50                     # Number of epochs
patience = 10                   # Epochs to wait for no observable improvement for early stopping of training
batch = 2                       # Number of images per batch
optimizer = 'Adam'              # Optimiser
verbose = True                  # Visual Representation of progress
dropout = 0.4                   # Dropout layers for Regularisation
learning_rate_start = 0.0001    # Initial learning rate
learning_rate_final = 0.00001   # Final learning rate
project = '/content/drive/MyDrive/Dissertation/TrainingResults/'    # Directory to save results
```

##### Yolo-v8 small model training

```python
model_s.train(data=data,
            epochs = epochs,
            patience = patience,
            batch = batch,
            optimizer = optimizer,
            verbose = verbose,
            dropout = dropout,
            lr0= learning_rate_start,
            lrf= learning_rate_final,
            name = 'yolov8-s',
            project = project)
```
